```python
temp = float(input("Enter the temperature"))
unit = input("Enter the unit (C,F,K): ").upper()
convert_to = input("Enter the unit you want to convert (C,F,K)").upper()

if unit == convert_to:
    print("You have entered the same unit !!!")

elif unit == "C":
    if convert_to == "F":
        print(temp * 9 / 5 + 32)
    elif convert_to == "K":
        print(temp + 273.15)

elif unit == "F":
    if convert_to == "C":
        print((temp - 32) * 5 / 9)
    elif convert_to == "K":
        print(((temp - 32) * 5 / 9 + 273.15))

elif unit == "K":
    if convert_to == "C":
        print(temp - 273.15)
    elif convert_to == "F":
        print((temp - 273.15) * 9/5 + 32)

else:
    print("Error")
```
