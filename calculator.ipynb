```python
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

sign = input("Enter an operator (+, -, *, /,%,**): ")

if sign == "+":
    print(num1 + num2)
elif sign == "-":
    print(num1 - num2)
elif sign == "*":
    print(num1 * num2)
elif sign == "/":
    if num2 == 0:
     print("Cannot divide by zero.")
    else:
     print(num1 / num2)
elif sign == "%":
   if num2 == 0:
      print("Cannot divide by zero.")
   else:
      print(num1 % num2)
elif sign == "**":
   print(pow(num1, 2) + pow(num2, 2))
   

else:
   print("You have entered an unrecognized operator")
```
