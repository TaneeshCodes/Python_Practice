Conditional Expression: A one - line shortcut for the if-else statement (ternary operator). Print or assign one of the two values based on a condition. X if condition else Y

```python
num = int(input("Enter your number: "))
print("Even" if num % 2 == 0 else "Odd")
```

```python
a = int(input("Enter your number 1: "))
b = int(input("Enter your number 2: "))
max = print(a if a > b else b )
```

```python
age = int(input("Enter your age: "))
print("Adult" if age > 18 else "Child")
```
