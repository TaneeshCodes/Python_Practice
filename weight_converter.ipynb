```python
weight = float(input("Enter your weight: "))
unit = input("Current unit (K/L/G/O): ").upper()
convert_to = input("Convert to (K/L/G/O): ").upper()

if unit == convert_to:
    print("Same unit entered.")

elif unit == "K":
    if convert_to == "L":
        print(weight * 2.20462)
    elif convert_to == "G":
        print(weight * 1000)
    elif convert_to == "O":
        print(weight * 35.274)

elif unit == "L":
    if convert_to == "K":
        print(weight / 2.20462)
    elif convert_to == "G":
        print(weight * 453.592)
    elif convert_to == "O":
        print(weight * 16)

elif unit == "G":
    if convert_to == "K":
        print(weight / 1000)
    elif convert_to == "L":
        print(weight / 453.592)
    elif convert_to == "O":
        print(weight / 28.3495)

elif unit == "O":
    if convert_to == "K":
        print(weight / 35.274)
    elif convert_to == "L":
        print(weight / 16)
    elif convert_to == "G":
        print(weight * 28.3495)

else:
    print("Invalid unit.")
```

Pro Level Code 

```python
weight = float(input("Enter your weight: "))
unit = input("Current unit (K/L/G/O): ").upper()
convert_to = input("Convert to (K/L/G/O): ").upper()

conversion = {
    ("K", "L"): 2.20462,
    ("K", "G"): 1000,
    ("K", "O"): 35.274,

    ("L", "K"): 1 / 2.20462,
    ("L", "G"): 453.592,
    ("L", "O"): 16,

    ("G", "K"): 1 / 1000,
    ("G", "L"): 1 / 453.592,
    ("G", "O"): 1 / 28.3495,

    ("O", "K"): 1 / 35.274,
    ("O", "L"): 1 / 16,
    ("O", "G"): 28.3495
}

if unit == convert_to:
    print("Same unit entered.")

elif (unit, convert_to) in conversion:
    print(weight * conversion[(unit, convert_to)])

else:
    print("Invalid unit.")
```
