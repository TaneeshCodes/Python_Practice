Drivers License

```python
driver_age = int(input("Enter your age: "))
test = input("Did you pass the test: (Y / N)")

if driver_age >= 18 and test == "Y":
    print("Here's your driver license")

elif driver_age < 18 and test == "Y":
    print("How the hell did u even pass the exam")

elif driver_age >=18 and test == "N":
    print("Try next year lil bro")

else:
    print("You are not eligible")
```

College Admission Checker

```python
marks = float(input("Enter your marks: "))
entrance_exam = input("Did you clear the entrance exam: (Y / N)")
documents = input("Do you have all of your documents handy: (Y / N)")

if marks >= 75 and entrance_exam == "Y" and documents == "Y":
    print("Admission Granted")

elif marks <= 75 or entrance_exam == "Y" and documents == "Y":
    print("Your marks are low")

elif marks >= 75 or not entrance_exam == "N" and documents == "Y":
    print("Entrance not cleared")

elif marks >= 75 and entrance_exam == "Y" or not documents == "N":
    print("Bring your documents first")

else:
    print("error")
```
